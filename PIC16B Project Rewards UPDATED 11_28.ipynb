```python
from dataclasses import dataclass
from typing import Dict
```

```python
import json

def load_cards(path: str = "clean_cards.json") -> list[dict]:
    with open(path, "r", encoding="utf-8") as f:
        cards = json.load(f)
    return cards

cards = load_cards("clean_cards.json")
cards
```

```python
from numeric_values import value_card_numeric
from spend_profile import SpendProfile

# Get rid of duplicate cards
unique_cards = {}
for card in cards:
    unique_cards[card["name"]] = card

cards = list(unique_cards.values())

# Sample case
sp = SpendProfile(
    travel=300,
    groceries_dining=0,
    gas_utilities=150,
    other=200
)

results = []
for card in cards:
    name = card["name"]
    val = value_card_numeric(card, sp)
    results.append((name, val))

top10 = sorted(results, key=lambda x: x[1], reverse=True)[:10]

for name, val in top10:
    print(f"{name}: ${val:.2f}")
```
